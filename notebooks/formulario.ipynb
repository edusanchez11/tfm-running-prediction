```python
import pandas as pd
import numpy as np
```

```python
# leer archivo csv

formulario = pd.read_csv(r"C:\Users\eduar\OneDrive\Escritorio\Master_DataScience_AI\TFM. ‚ÄúPredicci√≥n del rendimiento en corredores mediante t√©cnicas de Machine Learning un enfoque aplicado al sector deportivo‚Äù\data\forms\respuestas_forms.csv")

# ver las primeras filas
formulario.head()
```

```python
# reemplazar nombres columnas

formulario = formulario.rename(columns={
    'Nombre (obligatorio):\nEste ser√° usado para unir tu formulario con los registros de Srava': 'Nombre',
    "  Edad (a√±os)  ": 'Edad',
    "  Sexo/G√©nero:  ": 'Sexo',
    "  Peso (kg)  ": 'Peso',
    "  Altura (cm)  ": 'Altura',
    "Nivel de experiencia en running:\nPrincipiante (menos de 1 a√±o entrenando regularmente)\nIntermedio (1‚Äì3 a√±os entrenando regularmente)\nAvanzado/Competitivo (m√°s de 3 a√±os entrenando, compito en pruebas)": 'Nivel_experiencia',
    "  Objetivo principal actual:" : 'Objetivo',
    "  D√≠as que entrenas por semana (aprox)  ": 'Dias_entrenamiento',
})

# quedarnos solo con las columnas que nos interesan

formulario = formulario[['Nombre', 'Edad', 'Sexo', 'Peso', 'Altura', 'Nivel_experiencia', 'Objetivo', 'Dias_entrenamiento']]
formulario.head(20)
```

```python
# guardar el Dataframe en un archivo csv

formulario.to_csv('formulario_limpio.csv', index=False)
```
